# 1、读入数据Fund

```python
# 读取 rank
def read_fund_rank():
    df = pd.read_excel('../JN_DataWarehouse/fund/基金排名.xlsx' ,index_col=False)
    df['基金代码'] = df['基金代码'].apply(lambda x:str(x).zfill(6))                                #6位字符补0显示代码
    df['涨幅'] = df['涨幅']*100
    del df['序号']
    return df

# 读取 fund
def read_fund():
    df = pd.read_excel('../JN_DataWarehouse/fund/基金.xlsx' ,index_col=False)
    df['基金代码'] = df['基金代码'].apply(lambda x:str(x).zfill(6)) 
    df = df.replace('---', np.NaN)
    df = df.replace('--', np.NaN)
    df['日增长率'] = df['日增长率']*100
    df['近1周'] = df['近1周']*100
    df['近1月'] = df['近1月']*100
    df['近3月'] = df['近3月']*100
    df['近6月'] = df['近6月']*100
    df['近1年'] = df['近1年']*100
    df['近2年'] = df['近2年']*100
    df['近3年'] = df['近3年']*100
    df['今年来'] = df['今年来']*100
    df['成立来'] = df['成立来']*100
    df['手续费'] = df['手续费']*100
    
    df = df.drop(labels=["自定义 ","天天fund"],axis=1)
    
    return df

# 保存 新排名
def to_new_fund(df):
    df = df.to_excel('../JN_DataWarehouse/fund/基金排名new.xlsx')
```

```python
%%time
df_fund_rank = read_fund_rank()
df_fund = read_fund()
```
