```python
# 读取year
def read_newyear():
    df = pd.read_csv('../JN_DataWarehouse/stock_analysis/TuShare/new_year.csv' ,index_col=False)
    df['ts_code'] = df['ts_code'].apply(str)
    del df['Unnamed: 0']
    return df
```

```python
df_stock_all = read_newyear()
```
