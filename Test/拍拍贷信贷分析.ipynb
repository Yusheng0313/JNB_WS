```python
import matplotlib.pyplot as plt  
import numpy as np

money = 500
interest = 15  # 年利率
interval = 6   # 期限6个月
# 每期的应收金额
Receivables = ( money * interest/100/12 * (1 +interest/100/12)**interval )/( (1 +interest/100/12)**interval -1 )
Receivables = (float(Receivables)//0.01)*(0.01)
n = np.arange(1, interval+1, 1)  # 取 1-6 期
def principle(interval):   # 定义第n期的本金函数
    return (Receivables -money*interest/100/12)*(1 +interest/100/12)**(interval-1)
plt.plot( n ,principle(n),'ro')
plt.title("")
plt.ylabel('Principle/Yuan')
plt.xlabel('Interval/month')
# 设置两个坐标轴的范围
plt.ylim(80.5,86.5)
plt.xlim(0.5, 6.5) 
plt.figure(figsize=(15,10))
plt.grid() #开启网格
print('1-6期中，应收本金变化图：')
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np
money = 1000
interest = 10  # 年利率
interval = 12  # 期限12个月
n = np.arange(1, interval+1, 1)  # 取 1-12 期
def XMing_Gain(interval):   # 定义小明期限为interval的总收益（散标）
    Receivables = ( money * interest/100/12 * (1 +interest/100/12)**interval )/( (1 +interest/100/12)**interval -1 )
    return ((Receivables)*interval//0.01)*(0.01)
def XBai_Gain(interval):   # 定义小明期限为interval的总收益（彩虹计划）
    Receivables = money*(1 +interest/100/12)**interval
    return ((Receivables)//0.01)*(0.01)
plt.plot( n ,XBai_Gain(n),'ro',label="XiaoBai",color="red")
plt.plot( n ,XMing_Gain(n),'ro',label="XiaoMing",color="blue")
plt.title("")
plt.ylabel('Total Gain/Yuan')
plt.xlabel('Interval/month')
plt.legend() # 开启图例
plt.grid() # 开启网格
print('12种期限下，小明（散标）与小白（彩虹计划）各自的总收益图：')
plt.show()
```
