```python
# -*- coding: utf-8 -*-
"""
Created on Sun Mar  21 19:42:02 2020

@author: yihao.he
"""

from bs4 import BeautifulSoup  
import pandas as pd
from tqdm import tqdm
import math
import requests  
import lxml
import re
import time


# æˆ¿æºé¢ç§¯-æ€»ä»·æ•£ç‚¹å›¾
# å„è¡Œæ”¿åŒºå‡ä»·
# å‡ä»·æœ€è´µçš„10ä¸ªåœ°æ®µ
# å‡ä»·æœ€è´µçš„10ä¸ªå°åŒº
# æˆ·å‹åˆ†å¸ƒ
# æ ‡é¢˜æ–‡æœ¬è¯äº‘å›¾


from pyecharts.charts import *
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
from jieba import posseg as psg                         #ä¸­æ–‡åˆ†è¯è§£é‡Š
import collections


# è¯»å…¥CSV å‘½åä¸ºdata
data = pd.read_csv('.\\JN_DataWarehouse\\foshan_lianjiafangyuan\\lianjia-foshanfangyuan.csv')
print('start:',data.head())


# æˆ¿æºé¢ç§¯-æ€»ä»·æ•£ç‚¹å›¾
# ğŸ’¥æœ€è´µçš„ä¸€å¥—æˆ¿æºæ˜¯ä½äºå®å®‰-æ›¦åŸçš„æˆ¿æºï¼Œæ€»ä»·8800Wï¼›

scatter = (Scatter(init_opts=opts.InitOpts(theme='dark'))
           .add_xaxis(data['hourseSize'])
           .add_yaxis("æˆ¿ä»·", data['total_price'])
           .set_series_opts(label_opts=opts.LabelOpts(is_show=False),
                           markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max", name="æœ€å¤§å€¼"),]))
           .set_global_opts(
               legend_opts=opts.LegendOpts(is_show=False),
               title_opts=opts.TitleOpts(title="ä½›å±±äºŒæ‰‹æˆ¿ æ€»ä»·-é¢ç§¯ æ•£ç‚¹å›¾"),
               xaxis_opts=opts.AxisOpts(
                   name='é¢ç§¯',
                   # è®¾ç½®åæ ‡è½´ä¸ºæ•°å€¼ç±»å‹
                   type_="value", 
                   # ä¸æ˜¾ç¤ºåˆ†å‰²çº¿
                   splitline_opts=opts.SplitLineOpts(is_show=False)),
               yaxis_opts=opts.AxisOpts(
                   name='æ€»ä»·',
                   name_location='middle',
                   # è®¾ç½®åæ ‡è½´ä¸ºæ•°å€¼ç±»å‹
                   type_="value",
                   # é»˜è®¤ä¸ºFalseè¡¨ç¤ºèµ·å§‹ä¸º0
                   is_scale=True,
                   splitline_opts=opts.SplitLineOpts(is_show=False),),
               visualmap_opts=opts.VisualMapOpts(is_show=True, type_='color', min_=100, max_=1000)
    ))

scatter.render_notebook()                      #ç›´æ¥åœ¨jupyter notebook æ˜¾ç¤º
#scatter.render('æ•£ç‚¹å›¾.html')                #ç”Ÿæˆhtmlæ–‡ä»¶





# å„è¡Œæ”¿åŒºå‡ä»·
# æœ€è´µçš„æ˜¯å—å±±åŒºï¼Œæ•´ä½“å‡ä»·9.2W/å¹³ç±³ï¼›
# æœ€ä¾¿å®œçš„æ˜¯åªå±±åŒºï¼Œå‡ä»·3.5W/å¹³ç±³ï¼›

temp = data.groupby(['area'])['unit_price'].mean().reset_index()
data_pair = [(row['area'], round(row['unit_price']/10000, 1)) for _, row in temp.iterrows()]

map_ = (Map(init_opts=opts.InitOpts(theme='dark'))
        .add("äºŒæ‰‹æˆ¿å‡ä»·", data_pair, 'ä½›å±±', is_roam=False)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="ä½›å±±å„è¡Œæ”¿åŒºäºŒæ‰‹æˆ¿å‡ä»·"),
            legend_opts=opts.LegendOpts(is_show=False),
            tooltip_opts=opts.TooltipOpts(formatter='{b}:{c}ä¸‡å…ƒ'),
            visualmap_opts=opts.VisualMapOpts(min_=0, max_=3)
        )
       )

        
map_.render_notebook()
#map_.render('åœ°ç†å›¾.html')





# å‡ä»·æœ€é«˜çš„10ä¸ªå°åŒº
# è¯¥å°åŒºå†…åœ¨å”®æˆ¿æºè‡³å°‘3å¥—æ‰ç»Ÿè®¡

temp = data.groupby(['community'])['unit_price'].agg(['mean', 'count']).reset_index()

# è¯¥å°åŒºå†…è‡³å°‘3å¥—åœ¨å”®æˆ¿æºæ‰ç»Ÿè®¡
data_pair = sorted([(row['community'], round(row['mean']/10000, 1)) if row['count']>=3 else (0, 0)
                    for _, row in temp.iterrows()], key=lambda x: x[1], reverse=True)[:10]

bar = (Bar(init_opts=opts.InitOpts(theme='dark'))
       .add_xaxis([x[0] for x in data_pair[::-1]])
       .add_yaxis('äºŒæ‰‹æˆ¿å‡ä»·', [x[1] for x in data_pair[::-1]])
       .set_series_opts(label_opts=opts.LabelOpts(is_show=True, 
                                                       position='insideRight',
                                                       font_style='italic'),
                            itemstyle_opts=opts.ItemStyleOpts(
                                color=JsCode("""new echarts.graphic.LinearGradient(1, 0, 0, 0, 
                                             [{
                                                 offset: 0,
                                                 color: 'rgb(0,206,209)'
                                             }, {
                                                 offset: 1,
                                                 color: 'rgb(218,165,32)'
                                             }])"""))
                            )
       .set_global_opts(
           title_opts=opts.TitleOpts(title="ä½›å±±äºŒæ‰‹æˆ¿å‡ä»·TOP 10å°åŒº"),
           legend_opts=opts.LegendOpts(is_show=False),
           tooltip_opts=opts.TooltipOpts(formatter='{b}:{c}ä¸‡å…ƒ'),
           xaxis_opts=opts.AxisOpts(max_=5),
       )
       .reversal_axis()
      )

bar.render_notebook()
#bar.render('æŸ±çŠ¶å›¾10å°åŒº.html')




# å‡ä»·æœ€é«˜çš„10ä¸ªåœ°æ®µ

temp = data.groupby(['position'])['unit_price'].mean().reset_index()
data_pair = sorted([(row['position'], round(row['unit_price']/10000, 1))
                    for _, row in temp.iterrows()], key=lambda x: x[1], reverse=True)[:10]

bar = (Bar(init_opts=opts.InitOpts(theme='dark'))
       .add_xaxis([x[0] for x in data_pair])
       .add_yaxis('äºŒæ‰‹æˆ¿å‡ä»·', [x[1] for x in data_pair])
       .set_series_opts(label_opts=opts.LabelOpts(is_show=True, font_style='italic'),
                            itemstyle_opts=opts.ItemStyleOpts(
                                color=JsCode("""new echarts.graphic.LinearGradient(0, 1, 0, 0, 
                                             [{
                                                 offset: 0,
                                                 color: 'rgb(0,206,209)'
                                             }, {
                                                 offset: 1,
                                                 color: 'rgb(218,165,32)'
                                             }])"""))
                            )
       .set_global_opts(
           title_opts=opts.TitleOpts(title="ä½›å±±äºŒæ‰‹æˆ¿å‡ä»·TOP 10åœ°æ®µ"),
           legend_opts=opts.LegendOpts(is_show=False),
           tooltip_opts=opts.TooltipOpts(formatter='{b}:{c}ä¸‡å…ƒ'))
      )

bar.render_notebook()
#bar.render('æŸ±çŠ¶å›¾10åœ°æ®µ.html')





# æˆ·å‹åˆ†å¸ƒ
# ä¸‰å®¤ä¾ç„¶æ˜¯ä¸»åŠ›ï¼›
# åœ¨ä½›å±±è¿™ç§å¯¸åœŸå¯¸é‡‘çš„åœ°æ–¹ï¼Œå°æˆ·å‹å æ¯”ä¹Ÿå¾ˆå¤šï¼›

temp = data.groupby(['hourseType'])['area'].count().reset_index()
data_pair = sorted([(row['hourseType'], row['area'])
                    for _, row in temp.iterrows()], key=lambda x: x[1], reverse=True)[:10]

pie = (Pie(init_opts=opts.InitOpts(theme='dark'))
       .add('', data_pair,
            radius=["30%", "75%"],
            rosetype="radius")
       .set_global_opts(title_opts=opts.TitleOpts(title="ä½›å±±äºŒæ‰‹æˆ¿ æˆ·å‹åˆ†å¸ƒ"),
                       legend_opts=opts.LegendOpts(is_show=False),)
       .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}%"))
      )

pie.render_notebook()
#pie.render('æˆ·å‹å æ¯”å›¾.html')




# è¯äº‘å›¾
# æ¥çœ‹çœ‹æˆ¿æºæ ‡é¢˜ä¸­å‡ºç°æœ€å¤šçš„è¯è¯­

word_list = []
stop_words = ['èŠ±å›­','ä¸šä¸»','å‡ºå”®']
string =  str(''.join([i for i in data['title'] if isinstance(i, str)]))


words = psg.cut(string)
for x in words:
    if len(x.word)==1:
        pass
    elif x.flag in ('m', 'x'):
        pass
    elif x.word in stop_words:
        pass
    else:
        word_list.append(x.word)


data_pair = collections.Counter(word_list).most_common(100)


wc = (WordCloud()
      .add("", data_pair, word_size_range=[20, 100], shape='triangle')
      .set_global_opts(title_opts=opts.TitleOpts(title="æˆ¿æºæè¿°è¯äº‘å›¾"))
    )

wc.render_notebook()
#wc.render('è¯äº‘å›¾.html')


```
